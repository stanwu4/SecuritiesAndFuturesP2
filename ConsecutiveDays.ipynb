```python
import pandas as pd
import numpy as np
import datetime
from datetime import timedelta
```

```python
#tasked to find for each combination, what is the first day you see in the week, what is the last day you see in the week
#and how many days consecutively are missing after the first date they are seen
df = pd.read_csv('C:/Users/culle/Documents/ratios/futures.csv')
```

```python
df.head()
```

```python
#determine the dates/week to include, make a new data frame
df['clientInsertReqTimestamp'] = pd.to_datetime(df['clientInsertReqTimestamp'])
latest_date = df['clientInsertReqTimestamp'].max()
start_date = latest_date - datetime.timedelta(days=6)
lastSevenDays = df.loc[df['clientInsertReqTimestamp'] >= start_date]
lastSevenDays.head()
```

```python
#remove the time from the date
lastSevenDays['clientInsertReqTimestamp'] = pd.to_datetime(lastSevenDays['clientInsertReqTimestamp']).dt.date
lastSevenDays.head()
```

```python
#remove unneccessary variables
del lastSevenDays['combOffsetFlag']
del lastSevenDays['direction']
del lastSevenDays['limitPrice']
del lastSevenDays['accountRef']
del lastSevenDays['userId']
del lastSevenDays['clientId']
```

```python
#sort the earliest dates
subsetCols = ['clientIp','exchangeIp','counterIp','instrumentId']
lastSevenDays['earliestDate'] = lastSevenDays.groupby(subsetCols)['clientInsertReqTimestamp'].transform('min')
lastSevenDays.head()
```

```python
# the latest date
subsetCols = ['clientIp','exchangeIp','counterIp','instrumentId']
lastSevenDays['lastestDate'] = lastSevenDays.groupby(subsetCols)['clientInsertReqTimestamp'].transform('max')
lastSevenDays.head()
```

```python
#get the difference between the current date, and the earliest date
lastSevenDays['daysMissing'] = lastSevenDays['clientInsertReqTimestamp'] - lastSevenDays['earliestDate']
lastSevenDays['daysMissing'] = pd.to_timedelta(lastSevenDays['daysMissing']).dt.days
lastSevenDays.head()
```

```python
#delete unneccessary columns
del lastSevenDays['Unnamed: 0']
del lastSevenDays['exchangeRspLatency']
del lastSevenDays['exchangeRtnLatency']
```

```python
#drop the trades that happened on the same date, then sort the daysMissing column by the minimum,
#then you can drop duplicates to get the next date that you see this group being used

mask = (lastSevenDays.iloc[:, 7] != 0) | (lastSevenDays.iloc[:, 5] == lastSevenDays.iloc[:, 6])

# Filter the DataFrame using the boolean mask
lastSevenDays = lastSevenDays[mask].reset_index(drop=True)
```

```python
lastSevenDays.tail()
```

```python
#group by subset, sort by having the minimum first
subsetCols = ['clientIp','exchangeIp','counterIp','instrumentId']
lastSevenDays['daysMissing'] = lastSevenDays.groupby(subsetCols)['daysMissing'].transform('min')
lastSevenDays.tail()
```

```python
#delete duplicates
uniqueLastSevenDays = lastSevenDays.drop_duplicates(subset = ['clientIp','exchangeIp','counterIp','instrumentId'])
```
