```python
#Determine the date that was from max date - 7 days ago
df['clientInsertReqTimestamp'] = pd.to_datetime(df['clientInsertReqTimestamp'])
latest_date = df['clientInsertReqTimestamp'].max()
start_date = latest_date - datetime.timedelta(days=6)
lastSevenDays = df.loc[df['clientInsertReqTimestamp'] >= start_date]
lastSevenDays.head()
```

```python
#amount of data in the last 7 days
print(len(lastSevenDays))
```

```python
#determine that weight will be even weighting, 25% for all categories
```

```python
#delete unnecessary rows
del lastSevenDays['combOffsetFlag']
del lastSevenDays['direction']
del lastSevenDays['limitPrice']
del lastSevenDays['accountRef']
del lastSevenDays['userId']
del lastSevenDays['clientId']
```

```python
lastSevenDays.head()
```

```python
#find out the means of the latencies, for each variable
lastSevenDays['totalLatency'] = lastSevenDays['exchangeRspLatency'] + lastSevenDays['exchangeRtnLatency']
```

```python
counterAvg = pd.DataFrame(lastSevenDays[["counterIp","totalLatency"]].groupby("counterIp").mean())
counterAvg.head()
```

```python
clientAvg = pd.DataFrame(lastSevenDays[["clientIp","totalLatency"]].groupby("clientIp").mean())
clientAvg.head()
```

```python
exchangeAvg = pd.DataFrame(lastSevenDays[["exchangeIp","totalLatency"]].groupby("exchangeIp").mean())
exchangeAvg.head()
```

```python
#using hashtable to store values
instrumentAvg = pd.DataFrame(lastSevenDays[["instrumentId","totalLatency"]].groupby("instrumentId").mean())
instrumentAvg.head()
```

```python
#create a new column of the weighted averages in the lastSevenDays dataframe, 
#then take values of each, sum them up in the lastSevenDays dataframe
#this one takes a while to run
lastSevenDays['weightedLatency'] = 0

for i in range(len(lastSevenDays)):
    exchange = lastSevenDays.iloc[i,6]
    instr = lastSevenDays.iloc[i,7]
    client = lastSevenDays.iloc[i,4]
    counter = lastSevenDays.iloc[i,5]
    lastSevenDays.iloc[i,9] = (instrumentAvg.loc[instr,'totalLatency'] * 0.25 + exchangeAvg.loc[exchange,'totalLatency'] * 0.25 
    + clientAvg.loc[client,'totalLatency']* 0.25 + counterAvg.loc[counter,'totalLatency'] * 0.25)
    

```

```python
lastSevenDays.head()
```

```python
#divide the weighted averge by 1000000
lastSevenDays['weightedLatency'] /= 1000000
```

```python
#sort by weighted averages of latencies
lastSevenDays = lastSevenDays.sort_values(by = 'weightedLatency')
```
