```python
import pandas as pd
import numpy as np
import datetime
from datetime import timedelta
```

```python
df = pd.read_csv('C:/Users/culle/Documents/ratios/futures.csv')
```

```python
df.head()
```

```python
#Determine the date that was from max date - 7 days ago
df['clientInsertReqTimestamp'] = pd.to_datetime(df['clientInsertReqTimestamp'])
latest_date = df['clientInsertReqTimestamp'].max()
start_date = latest_date - datetime.timedelta(days=6)
lastSevenDays = df.loc[df['clientInsertReqTimestamp'] >= start_date]
lastSevenDays.head()
```

```python
#remove the time from datetime
lastSevenDays['clientInsertReqTimestamp'] = pd.to_datetime(lastSevenDays['clientInsertReqTimestamp']).dt.date
lastSevenDays.head()
```

```python
#delete unnecessary rows
del lastSevenDays['combOffsetFlag']
del lastSevenDays['direction']
del lastSevenDays['limitPrice']
del lastSevenDays['accountRef']
del lastSevenDays['userId']
del lastSevenDays['clientId']
```

```python
lastSevenDays.head()
```

```python
#find out the means of the latencies, for each variable
lastSevenDays['totalLatency'] = lastSevenDays['exchangeRspLatency'] + lastSevenDays['exchangeRtnLatency']
lastSevenDays = lastSevenDays.copy()
```

```python
#created weighted Latencies for all entries in the last 7 days, using hash table to store values of expoential weight
lastSevenDays['weightedLatency'] = 0
weights = {
    '09': 0,
    '08': 1,
    '07': 2,
    '06': 3,
    '05': 4,
    '04': 5,
}

lastSevenDays['clientInsertReqTimestamp'] = pd.to_datetime(lastSevenDays['clientInsertReqTimestamp'])
lastSevenDays['weightedLatency'] = lastSevenDays.iloc[:, 8] * pow(0.8, lastSevenDays.iloc[:, 1].dt.strftime("%d").map(weights).fillna(6))
```

```python
#creating another df, that is only unique
#clone a new df and delete duplicates, remove excess columns
uniqueLastSevenDays = lastSevenDays
uniqueLastSevenDays = uniqueLastSevenDays.drop_duplicates(subset = ['clientIp','exchangeIp','counterIp','instrumentId'])
del uniqueLastSevenDays['Unnamed: 0']
del uniqueLastSevenDays['exchangeRspLatency']
del uniqueLastSevenDays['exchangeRtnLatency']
uniqueLastSevenDays.head()
```

```python
lastSevenDays_grouped = lastSevenDays.groupby(['clientIp', 'counterIp', 'exchangeIp', 'instrumentId'])
subset = lastSevenDays_grouped['weightedLatency'].sum() / 1000000
uniqueLastSevenDays = uniqueLastSevenDays.merge(subset, on=['clientIp', 'counterIp', 'exchangeIp', 'instrumentId'], how='left')
```

```python
uniqueLastSevenDays.rename(columns={'weightedLatency_y':'weightedAndSmoothed'},inplace = True)
uniqueLastSevenDays.head()
```

```python
uniqueLastSevenDays.sort_values(by = 'weightedAndSmoothed')
uniqueLastSevenDays.head()
```

```python
#create a new df to store the medians of the top 5 for each of the last 7 days
numOccur_df = pd.DataFrame("none",index = range(5),columns=['D1', 'D2', 'D3', 'D4', 'D5', 'D6', 'D7'])
```

```python
result_df = uniqueLastSevenDays.head(5)
del result_df['clientInsertReqTimestamp']
del result_df['totalLatency']
del result_df['weightedLatency_x']
result_df.head()
```

```python
#creating the medians of the top 5 latencies for each day
for i in range(5):
    startDate = datetime.datetime(2022,12,3) 
    client = result_df.iloc[i,0]
    counter = result_df.iloc[i,1]
    exchange = result_df.iloc[i,2]
    instrument = result_df.iloc[i,3]
    for j in range(7):
        subset = lastSevenDays[(lastSevenDays['clientIp'] == client) & 
                       (lastSevenDays['counterIp'] == counter) & 
                       (lastSevenDays['exchangeIp'] == exchange) & 
                       (lastSevenDays['instrumentId'] == instrument) &
                       (lastSevenDays['clientInsertReqTimestamp'].dt.date  == startDate.date())]
        latency = subset['totalLatency'].median() / 1000000
        if latency == 0:
            numOccur_df.iloc[i,j] = "NaN"
        else:
            numOccur_df.iloc[i,j] = latency
            
        startDate += timedelta(days = 1)
        
numOccur_df.head()
```

```python
#change ratio
result_df['changeRatio'] = 0
for i in range(5):
    client = result_df.iloc[i,0]
    counter = result_df.iloc[i,1]
    exchange = result_df.iloc[i,2]
    instrument = result_df.iloc[i,3]
    subset = lastSevenDays[(lastSevenDays['clientIp'] == client) & 
                       (lastSevenDays['counterIp'] == counter) & 
                       (lastSevenDays['exchangeIp'] == exchange) & 
                       (lastSevenDays['instrumentId'] == instrument)]
    max = subset['totalLatency'].max() / 1000000
    min = subset['totalLatency'].min() / 1000000
    ratio = (max - min) / min
    result_df.iloc[i,5] = ratio
```

```python
#efficiency ratio
result_df['effRatio'] = 0
for i in range(5):
    client = result_df.iloc[i,0]
    counter = result_df.iloc[i,1]
    exchange = result_df.iloc[i,2]
    instrument = result_df.iloc[i,3]
    subset = lastSevenDays[(lastSevenDays['clientIp'] == client) & 
                       (lastSevenDays['counterIp'] == counter) & 
                       (lastSevenDays['exchangeIp'] == exchange) & 
                       (lastSevenDays['instrumentId'] == instrument)]
    ratio = subset['clientInsertReqTimestamp'].nunique()
    result_df.iloc[i,6] = ratio/7
```
